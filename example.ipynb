## Load the dataset

```python
import json
data = json.load(open('data.json'))
```

## Analyze the dataset

```python
## Print the first 10 objects
for i in range(10):
    print(data[i])
```

### Filter Dining Data

```python
dining_data = []
for object in data:
    if 'dining' in object['metadata']['source']:
        dining_data.append(object['metadata']['source'])

print(dining_data)
```

### Add the key: value to the data 

```python
for object in data:
    if object['metadata']['source'] in dining_data:
        object['metadata']['group'] = 'dining'
```

```python
# Check if the group is added
for object in data:
    if 'group' in object['metadata']:
        print(object['metadata'])
```
